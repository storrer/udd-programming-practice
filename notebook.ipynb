```python
word_list = "sowpods.txt"
```

```python
import os
```

```python
with open(word_list, 'r') as file:
    words = file.readlines()
print([word.strip() for word in words[0:4]])
```

```python
matching_words = []
for word in words:
    word = word.strip()
    if all(letter not in 'XYOI' for letter in word):
        matching_words.append(word)
print(matching_words)
```
